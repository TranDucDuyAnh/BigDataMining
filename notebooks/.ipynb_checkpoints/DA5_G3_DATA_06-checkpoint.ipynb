```python
import os
from pathlib import Path

base_dir = "/content/drive/MyDrive/DA5-G3-BigData"

# Input từ task 05
cleaned_txt_dir = os.path.join(base_dir, "data/processed/cleaned_text/cleaned_txt")

# Output task spaCy
output_dir = os.path.join(base_dir, "data/processed/spacy_pipeline")

token_json_dir = os.path.join(output_dir, "tokenized_json")
token_csv_dir = os.path.join(output_dir, "tokenized_csv")
parsed_txt_dir = os.path.join(output_dir, "parsed_txt")
summary_csv_dir = os.path.join(output_dir, "processed_csv")

os.makedirs(token_json_dir, exist_ok=True)
os.makedirs(token_csv_dir, exist_ok=True)
os.makedirs(parsed_txt_dir, exist_ok=True)
os.makedirs(summary_csv_dir, exist_ok=True)

print("Input cleaned text:", cleaned_txt_dir)
print("Output spaCy:", output_dir)
```

```python
cleaned_txt_files = list(Path(cleaned_txt_dir).rglob("*.txt"))

print("Số file TXT sạch tìm thấy:", len(cleaned_txt_files))

for file in cleaned_txt_files[:10]:
    print(file)
```

```python
import os
import json
import pandas as pd
from tqdm import tqdm
from pathlib import Path
import spacy


# Load model tiếng Anh
nlp = spacy.load("en_core_web_sm")

# Tăng giới hạn độ dài văn bản nếu báo cáo tài chính dài
nlp.max_length = 5_000_000


def split_long_text(text: str, max_chars: int = 100000) -> list:
    """
    Chia văn bản dài thành nhiều đoạn nhỏ để spaCy xử lý ổn định hơn.
    """
    chunks = []

    start = 0
    while start < len(text):
        end = start + max_chars

        # Cố gắng cắt tại dấu xuống dòng hoặc dấu chấm để không cắt giữa câu
        if end < len(text):
            newline_pos = text.rfind("\n", start, end)
            period_pos = text.rfind(".", start, end)

            cut_pos = max(newline_pos, period_pos)

            if cut_pos > start:
                end = cut_pos + 1

        chunks.append(text[start:end])
        start = end

    return chunks


def process_text_with_spacy(text: str) -> dict:
    """
    Thực hiện:
    - Tokenization
    - POS Tagging
    - Dependency Parsing
    """

    chunks = split_long_text(text)

    all_tokens = []
    all_sentences = []

    token_id = 1
    sentence_id = 1

    for chunk_id, chunk in enumerate(chunks, start=1):
        doc = nlp(chunk)

        for sent in doc.sents:
            sentence_tokens = []

            for token in sent:
                if token.is_space:
                    continue

                token_data = {
                    "token_id": token_id,
                    "chunk_id": chunk_id,
                    "sentence_id": sentence_id,
                    "text": token.text,
                    "lemma": token.lemma_,
                    "pos": token.pos_,
                    "tag": token.tag_,
                    "dep": token.dep_,
                    "head_text": token.head.text,
                    "is_alpha": token.is_alpha,
                    "is_stop": token.is_stop,
                    "is_punct": token.is_punct
                }

                all_tokens.append(token_data)
                sentence_tokens.append(token.text)

                token_id += 1

            if sentence_tokens:
                all_sentences.append({
                    "sentence_id": sentence_id,
                    "chunk_id": chunk_id,
                    "sentence_text": " ".join(sentence_tokens)
                })

                sentence_id += 1

    result = {
        "total_chunks": len(chunks),
        "total_sentences": len(all_sentences),
        "total_tokens": len(all_tokens),
        "sentences": all_sentences,
        "tokens": all_tokens
    }

    return result


def save_parsed_txt(spacy_data: dict, output_path: str):
    """
    Lưu kết quả parsing ở dạng TXT dễ đọc.
    """

    with open(output_path, "w", encoding="utf-8") as f:
        f.write("TOKEN\tLEMMA\tPOS\tTAG\tDEP\tHEAD\n")

        for token in spacy_data["tokens"]:
            line = (
                f"{token['text']}\t"
                f"{token['lemma']}\t"
                f"{token['pos']}\t"
                f"{token['tag']}\t"
                f"{token['dep']}\t"
                f"{token['head_text']}\n"
            )
            f.write(line)


def process_spacy_pipeline(txt_files, token_json_dir, token_csv_dir, parsed_txt_dir):
    """
    Chạy spaCy cho toàn bộ file TXT sạch.
    """

    summary = []

    for file_path in tqdm(txt_files, desc="Đang chạy spaCy pipeline"):
        try:
            file_path = Path(file_path)

            with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
                text = f.read()

            spacy_data = process_text_with_spacy(text)

            safe_name = file_path.stem.replace(" ", "_")

            json_output_path = Path(token_json_dir) / f"{safe_name}_spacy.json"
            csv_output_path = Path(token_csv_dir) / f"{safe_name}_tokens.csv"
            txt_output_path = Path(parsed_txt_dir) / f"{safe_name}_parsed.txt"

            # Lưu JSON
            output_json = {
                "file_name": file_path.name,
                "source_path": str(file_path),
                "total_chunks": spacy_data["total_chunks"],
                "total_sentences": spacy_data["total_sentences"],
                "total_tokens": spacy_data["total_tokens"],
                "sentences": spacy_data["sentences"],
                "tokens": spacy_data["tokens"]
            }

            with open(json_output_path, "w", encoding="utf-8") as f:
                json.dump(output_json, f, ensure_ascii=False, indent=2)

            # Lưu CSV token
            token_df = pd.DataFrame(spacy_data["tokens"])
            token_df.to_csv(csv_output_path, index=False, encoding="utf-8-sig")

            # Lưu TXT dễ đọc
            save_parsed_txt(spacy_data, txt_output_path)

            summary.append({
                "file_name": file_path.name,
                "source_path": str(file_path),
                "json_output": str(json_output_path),
                "csv_output": str(csv_output_path),
                "txt_output": str(txt_output_path),
                "total_chunks": spacy_data["total_chunks"],
                "total_sentences": spacy_data["total_sentences"],
                "total_tokens": spacy_data["total_tokens"],
                "status": "success"
            })

        except Exception as e:
            summary.append({
                "file_name": str(file_path),
                "source_path": str(file_path),
                "json_output": "",
                "csv_output": "",
                "txt_output": "",
                "total_chunks": 0,
                "total_sentences": 0,
                "total_tokens": 0,
                "status": f"error: {str(e)}"
            })

    return summary
```

```python
test_files = cleaned_txt_files[:2]

summary = process_spacy_pipeline(
    txt_files=test_files,
    token_json_dir=token_json_dir,
    token_csv_dir=token_csv_dir,
    parsed_txt_dir=parsed_txt_dir
)

summary_df = pd.DataFrame(summary)
summary_df
```

```python
summary = process_spacy_pipeline(
    txt_files=cleaned_txt_files,
    token_json_dir=token_json_dir,
    token_csv_dir=token_csv_dir,
    parsed_txt_dir=parsed_txt_dir
)

summary_df = pd.DataFrame(summary)

output_summary_csv = os.path.join(summary_csv_dir, "spacy_pipeline_summary.csv")
summary_df.to_csv(output_summary_csv, index=False, encoding="utf-8-sig")

print("Hoàn tất pipeline spaCy.")
print("File CSV tổng hợp:")
print(output_summary_csv)

summary_df
```
